#Data visualization
###Univariate Analysis

```python
#Univariate analysis of numerical features in price dataframe 
def pricedf(param):
    price_df[param].plot.hist(bins=16)
    plt.xlabel(param)
    plt.xticks(rotation=90);
    plt.show()
attributes=['price_off_peak_var', 'price_peak_var', 'price_mid_peak_var', 'price_off_peak_fix',
       'price_off_peak_fix', 'price_peak_fix', 'price_mid_peak_fix']
for attribute in attributes:
    pricedf(attribute)
plt.show()
```

```python
#Univariate analysis of numerical features in client dataframe 
def clientdf(param):
    client_df[param].plot.hist(bins=16)
    plt.xlabel(param)
    plt.xticks(rotation=90);
    plt.show()
attributes=['cons_12m', 'cons_gas_12m', 'cons_last_month', 'forecast_cons_12m', 'forecast_cons_12m', 'forecast_cons_year', 'forecast_discount_energy',
       'forecast_meter_rent_12m', 'forecast_price_energy_off_peak', 'forecast_price_energy_peak',
       'forecast_price_pow_off_peak', 'imp_cons', 'margin_gross_pow_ele',
       'margin_net_pow_ele', 'nb_prod_act', 'net_margin', 'num_years_antig',
       'pow_max', 'churn']
for attribute in attributes:
    clientdf(attribute)
plt.show()
```

#Correlation Between price sesitivity and churn

```python
#verifying if the price sensitivity is correlated with churn throught boxplots
def correl(param):
    sns.boxplot(y=price_df[param],x=client_df.churn)
    plt.xticks(rotation=90);
    plt.show()
attributes=[ 'price_peak_fix', 'price_peak_var', 'price_mid_peak_var', 'price_off_peak_fix', 'price_mid_peak_fix']
for attribute in attributes:
    correl(attribute)
plt.show()
```

#Analysis

####The boxplots above show us clearly that exist a correlation between price sensitivity and churn and the scatterplots bellow show us the same behaviour, So evidently exist a cerrelation between these features 

```python
#verifying if the price sensitivity is correlated with churn throught scatterplots
def correl(param):
    sns.scatterplot(y=price_df[param],x=client_df.churn)
    plt.xticks(rotation=90);
    plt.show()
attributes=[ 'price_peak_fix', 'price_peak_var', 'price_mid_peak_var', 'price_off_peak_fix', 'price_mid_peak_fix']
for attribute in attributes:
    correl(attribute)
plt.show()
```
